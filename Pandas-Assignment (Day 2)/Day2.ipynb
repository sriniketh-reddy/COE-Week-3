```python
#to find the rides started in january
df["START_DATE*"] = pd.to_datetime(df["START_DATE*"],errors="ignore")
df["END_DATE*"] = pd.to_datetime(df["END_DATE*"],errors="ignore")
df[(df["START_DATE*"].dt.month==1)].reset_index(drop=True)
```

```python
df[(df["START_DATE*"].dt.month==1) & (df["START_DATE*"].dt.day==2)].reset_index(drop=True)
```

```python
df[(df["START_DATE*"].dt.month==1) & (df["START_DATE*"].dt.year==2016) & (df["START*"]=="Cary")].reset_index(drop=True)
```
