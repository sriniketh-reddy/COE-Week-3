**Load the Dataset**

**Calculate the total sales for each store across all months.**

```python
ss["Total Sales"]=ss.iloc[:,2:14].sum(axis=1)
ss
```

**Find the average sales for each month across all stores**

```python
ss["Average Sales"] = ss["Total Sales"]/12
ss
```

**Identify the store with the highest total sales.**

```python
ss.sort_values("Total Sales",ascending=False).iloc[:1]
```

**Calculate the total sales for each city.**

```python
ss.groupby("city")["Total Sales"].agg("sum")
```
