# Assignment on Store_sales dataset

**Load the Dataset**

```python
import pandas as pd
ss = pd.read_csv("/store_sales.csv")
ss
```

**Calculate the total sales for each store across all months.**

```python
ss["Total Sales"]=ss.iloc[:,2:14].sum(axis=1)
ss
```

**Find the average sales for each month across all stores**

```python
ss["Average Sales"] = ss["Total Sales"]/12
ss
```

**Identify the store with the highest total sales.**

```python
ss.sort_values("Total Sales",ascending=False).iloc[:1]
```

**Calculate the total sales for each city.**

```python
ss.groupby("city")["Total Sales"].agg("sum")
```

**List stores with total sales greater than 200**

```python
ss[ss["Total Sales"]>200]
```

**Which month had the highest average sales across all stores?**


```python
months_average = ss.iloc[:,2:14].mean()
months_average
```

```python
months_average.idxmax(),months_average.max()
```

**Which city generated the most revenue?**

```python
city_sales = ss.groupby("city")["Total Sales"].agg("sum")
city_sales.idxmax(),"Total Sales: "+str(city_sales.max())
```
